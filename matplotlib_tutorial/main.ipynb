```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph 

```python
# Basic Graph
x = [1,2,3,4,5,6]
y = [4,5,6,7,8,9]

x2 = np.arange(0,4,0.5)
print(x2)

plt.plot(x,y,label = 'x',color = 'blue',marker = 'o')
plt.plot(x2,x2**2, color = 'red',marker = 'o')
plt.title("First Basic Graph", fontdict = {'fontsize': 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
# plt.xticks([0,1,2,3,4])
# plt.xticks([0,2,4,5,8,10])
plt.legend()
plt.savefig('graph.png',dpi=300)
plt.show()
```

### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]
# plt.bar(labels,values)
bars = plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))


plt.show()
```

### Real World Example

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(6,4))
plt.plot(gas.Year,gas.USA,'b.-',label='USA')
plt.plot(gas.Year,gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year,gas["South Korea"],'g.-',label='South Korea')

plt.title("USA vs Canada Gas Prices", fontdict = {'fontsize': 20})
plt.ylabel("Dollars/Gallon")
plt.legend()

```
